```python
def sorted(arr):
    srt = True
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            srt = False
            break
    return srt

def bubble_sort(arr):
    count = 0
    sorted = False
    i = 0
    while not sorted:
        sorted = True
        for j in range(len(arr) - 1 - i):
            count += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                sorted = False
        i += 1
    return arr, count
```

```python
def insertion_sort(array):
    cnt = 0
    for i in range(1, len(array)):
        for j in range(i-1, -1, -1):
            cnt += 1
            if array[j+1] < array[j]:
                array[j+1], array[j] = array[j], array[j+1]
    return array, cnt

a = [random.randint(0, 10) for _ in range(10)]
print(a)
s, c = insertion_sort(a)
print(c)
    
        
```

```python
def merge(arr1, arr2):
        i = 0
        j = 0
        new_arr = []
        count = 0
        while i < len(arr1) and j < len(arr2):
            count += 1
            if arr1[i] > arr2[j]:
                new_arr.append(arr2[j])
                j += 1
            else:
                new_arr.append(arr1[i])
                i += 1
        if i == len(arr1):
            new_arr += arr2[j:]
        elif j == len(arr2):
            new_arr += arr1[i:]
        count += 1
        return new_arr, count


def merge_sort(array):
    count = 0
    if len(array) > 1:
        pop = len(array) // 2
        left, cl = merge_sort(array[:pop])
        right, cr = merge_sort(array[pop:])
        array, c = merge(left, right)
        count += cl
        count += cr
        count += c
    return array, count



            
```

```python
def quick_sort(array, left, right):
    if right - left > 1:
        i_pivot = random.randint(left, right)
        pivot = array[i_pivot]
        array[right], array[i_pivot] = array[i_pivot], array[right]
        i_pivot = right
        right = right - 1
        # print('before')
        while right - left != -1:
            while array[left] <= pivot and right - left != -1:
                left += 1
                # print(left)
            while array[right] >= pivot and right - left != -1:
                right += -1
            # print('after')
            if right < left:
                array[i_pivot], array[left] = array[left], array[i_pivot]
            else:
                array[left], array[right] = array[right], array[left]
        quick_sort(array, 0, right)
        quick_sort(array, left, i_pivot)
    return array
```
